# Support Vector Machines #

Support Vector Machines are *linear classifiers* that find the "widest margin" between the  classes. 


     
**Why better than k-nearest neighbors?**  Computational ease . . . .

**Why worse than k-nearest neighbors?**  It's a *linear classifier* . . . . 

But . . . there is something called the 'kernel trick'. 

## Example ##

```python
myX2, myy2 = skd.make_circles(n_samples = 250, noise=.19, factor = .5)
```

```python
newX2 = np.transpose(myX2)
sns.scatterplot(x = newX2[0], y = newX2[1], hue = myy2)
```

```python
train_X, test_X, train_y, test_y = train_test_split(myX2, myy2, test_size=0.25)
```

```python
from sklearn.svm import SVC
mysvm = SVC(kernel='linear')
mysvm.fit(train_X, train_y)
```

```python
predicted = mysvm.predict(test_X)
```

```python
confusion_matrix(predicted, test_y)
```

### Using a Radial Basis Function kernel ###

The idea is to compose this function with a "kernel" function of the form:

$$
K(x,x') = exp\left( - {\|x - x'\|^2 \over 2 \sigma^2}\right)
$$

```python
mysvm = SVC(kernel='rbf')  # The default!
mysvm.fit(train_X, train_y)
```

```python
predicted = mysvm.predict(test_X)
```

```python
confusion_matrix(predicted, test_y)
```
