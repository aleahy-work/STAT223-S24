## How to plot a curve over a scatter plot ##

SKLearn has ways of generating [toy datasets](https://scikit-learn.org/stable/datasets/sample_generators.html). We'll use this to generate an interesting scatter plot.  As an example, we want to show how to plot this with a line (though an arbitrary curve will also work). 

```python
myX, myy = skd.make_blobs(centers=2,cluster_std=3.0)
```

```python
myX
```

```python
myy
```

```python
newX = np.transpose(myX)  # We need to transpose the data to fit it into a scatter plot
```

```python
linex = np.linspace(0,14,10)  # Plot a line first
```

```python
liney = 10 - 1.4*linex
```

```python
plt.plot(linex,liney)
```

```python
sns.scatterplot(x = newX[0] , y = newX[1], hue = myy)
plt.plot(linex,liney)
plt.show()
```
